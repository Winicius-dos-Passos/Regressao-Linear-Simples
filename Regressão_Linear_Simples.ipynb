```python
# Importação das bibliotecas, sklearn para criar o modelo de regressão e yellowbrick para visualização de residuai
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

```python
# Carregamento do dataset
df = pd.read_csv('cars.csv')
df.shape
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
# Definição das variavéis X e Y, X = distância é a variável independente e Y = velocidade é a variável dependente
X = df.iloc[:, 1].values
y = df.iloc[:, 0].values
X
```

```python
# Cálculo da correlação entre X e Y
correlacao = np.corrcoef(X, y)
correlacao
```

```python
# Formato de matriz com uma coluna a mais
X = X.reshape(-1, 1)
# Criação do modelo e treinamento (fit indica que o treinamento deve ser executado)
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Visualização dos coeficientes
modelo.intercept_
```

```python
# inclinação
modelo.coef_
```

```python
# Geração do gráfico com os pontos reais e as previsões
plt.scatter(X, y)
plt.plot(X, modelo.predict(X), color='red')
plt.title('Regressão Linear Simples')
plt.xlabel('Distância')
plt.ylabel('Velocidade')
plt.show()
```

```python
# Previsão utilizando função do sklearn
modelo.predict([[22]])
```

```python
# Gráfico para plotar os residuais
visualizador = ResidualsPlot(modelo)
visualizador.fit(X, y)
visualizador.show();
```
